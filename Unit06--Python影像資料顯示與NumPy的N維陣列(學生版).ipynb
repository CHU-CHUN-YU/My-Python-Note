###### NumPy : Base N-dimensional array package
* ndarray.ndim<br>
    the number of axes (dimensions) of the array. In the Python world, the number of dimensions is referred to as rank.
* ndarray.shape<br>
    the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the rank, or number of dimensions, ndim.
* ndarray.size<br>
    the total number of elements of the array. This is equal to the product of the elements of shape.
* ndarray.dtype<br>
    an object describing the type of the elements in the array. One can create or specify dtype’s using standard Python types. Additionally NumPy provides types of its own. numpy.int32, numpy.int16, and numpy.float64 are some examples.
* ndarray.itemsize<br>
    the size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.
* ndarray.data<br>
    the buffer containing the actual elements of the array. Normally, we won’t need to use this attribute because we will access the elements in an array using indexing facilities.

<font color=#0000FF>彩色圖(RGB三維) 轉成 灰階圖(二維)</font>
* [[OpenCV] 轉換影像為灰階 (Transform Image to Gray Level)](https://cg2010studio.com/2011/06/06/opencv-%E8%BD%89%E6%8F%9B%E5%BD%B1%E5%83%8F%E7%82%BA%E7%81%B0%E9%9A%8E-transform-image-to-gray-level/)
* <font color=#666666>Gray</font> = *0.299* * <font color=#FF0000>Red</font> + *0.587* * <font color=#00FF00>Green</font> + *0.114* * <font color=#0000FF>Blue</font>

```python
imgFile = r"E:\annie\3下課程\MyPython\PyImages\神力女超人.jpg"
img = plt.imread(imgFile)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
imgFile = r"E:\annie\3下課程\MyPython\PyImages\神力女超人.jpg"
img = plt.imread(imgFile)
print(img.shape)

# gray=0.299 * Red + 0.587 * Green + 0.114 * Blue
img2 = img[:,:,0]*0.299 + img[:,:,1]*0.587 + img[:,:,2]*0.114

print(img2.shape)
plt.imshow(img2, cmap="gray")  # cmap : color map
plt.show()
```

```python
def showColor2Gray(imgFile) :
    img = plt.imread(imgFile)
    img2 = img[:,:,0]*0.299 + img[:,:,1]*0.587 + img[:,:,2]*0.114
    print(img2.shape)
    plt.imshow(img2, cmap="gray")  # cmap : color map
    plt.show()
```

<font color=#0000FF>np.array的運算</font>

```python
x = np.array( ( (3,1)  , (2,5)  ) )
y = np.array( [ [5,7]  , [3,2]  ] )
print(x)
print(y)
print(type(x))
print(type(y))
print(x*y) #直接乘
```

```python
x = np.array( ( (3,1)  , (2,5)  ) )
y = np.array( [ [5,7]  , [3,2]  ] )
print(x)
print(y)
print(type(x))
print(type(y))
```

```python
print(x+y)
print(x-y)
print(x*y)  #直接乘
print(x/y)  #直接除
```

<font color=#0000FF>matrix的運算</font>

```python
x = np.matrix( ( (3,1)  , (2,5)  ) )
y = np.matrix( [ [5,7]  , [3,2]  ] )
print(type(x))
print(type(y))
print(x)
print(y)
```

```python
print(x+y)
print(x-y)
print(x*y)   # 特別注意 矩陣的乘法 不是直接乘
print(x/y)   #直接除
```

```python
a = np.matrix( ( (3,1,2)  , (2,5,1)  ) )
b = np.matrix( [ [5,7,2]  , [3,2,6]  ] )
print(a.shape)
print(b.shape)
```

```python
print(b)
print(b.shape)
print(b.T) #transform
print(b.T.shape)
```

```python
print(a * b.T)
```

```python
print(a * b.getT())  # b.T=b.getT()=b.transpose()
```

```python
print(a * b.transpose()) #transform
```

```python
x = np.matrix( ( (3,1)  , (2,5)  ) )
print(x)
print(np.diag(x))    # 對角線
print(np.trace(x))   # 對角線相加
```

```python
x = np.matrix( ( (3,1)  , (2,5)  ) )
eigvalue, eigvector = LA.eig(x) #loading

print(eigvalue)
print(eigvector)
```
