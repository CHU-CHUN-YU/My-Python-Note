### <font color=#0000FF>只有給定一個數列就是 y </font>

```python
np.random.seed(123) #set seed
plt.figure(figsize=(8,6))

y1 = np.random.choice(range(21), 5, replace=False)
y2 = np.random.choice(range(21), 5, replace=False)

plt.plot(y1,'r^:') #color/maker/linestyle 都放這
plt.plot(y2,'bv')

plt.show()
```

```python
plt.figure(figsize=(8,4))

x = np.linspace(-2*np.pi,2*np.pi,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'r-', label=r'$\sin(\theta)$')
plt.plot(x, y2, 'b-', label=r'$\cos(\theta)$')

plt.legend()

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           labels=[r'$-2\pi$',r'$-\pi$',0,r'$\pi$',r'$2\pi$'])

plt.hlines(y=0,xmin=-4,xmax=4)
plt.axvline(x=0,ymin=0.2,ymax=0.6) #ax 是百分比 沒有ax是絕對位置

plt.show()
```

```python
plt.figure(figsize=(8,4))

x = np.linspace(-2*np.pi,2*np.pi,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'r-', label=r'$\sin(\theta)$')
plt.plot(x, y2, 'b-', label=r'$\cos(\theta)$')
plt.legend()

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],labels=[r'$-2\pi$',r'$-\pi$',0,r'$\pi$',r'$2\pi$'])

#以下 plt.hlines or plt.vlines 最大、最小值是採用實際數值
plt.hlines(y=0,xmin=-4,xmax=4,color='k',linewidth=0.4,linestyle='-.')

#以下 plt.axhlines or plt.axvlines 最大、最小值是採用百分比
plt.axvline(x=0,ymin=0.2,ymax=0.6,color='r',linewidth=0.8,linestyle='--')

plt.show()
```

投100次骰子

```python
from random import randint

def rollDice():
    numSides = 6
    pointList = []
    rateList = []
    #
    for i in range(1000):
        pointList.append(randint(1, numSides))
    #
    # ------------------------------------
    # 試行1000次丟擲骰子的過程，randint(1, 6)
    # 回傳1~6(包括1,6本身)的隨機亂數值
    # ------------------------------------      
    # print(randint(1, numSides), end=",")
    # ------------------------------------
    #
    for pt in range(1, numSides+1):
        pointTimes = pointList.count(pt)
        rateList.append(pointTimes)
    # -------------------------------------
    # pt代表1~6的點數，利用List列表的count方法
    # 計算pt總出現的次數並存放於pointTimes，當
    # pt=1，便是計算骰子出現點數1時所有的次數，
    # 先存放在pointTimes，接續轉存於rateList
    # 這個列表中(該列表用來紀錄1~6點數各所出現的
    # 總次數
    # -------------------------------------
    #
    print(rateList)
    barGh = plt.bar([1,2,3,4,5,6],rateList)
   # barGh.title ("Result of rolling Dice 1000 times")
    barGh.x_labels = ["1", "2", "3", "4", "5", "6"]
    barGh.x_title = "Result"
    barGh.y_title = "Frequency of Result"
    #barGh.add("D6", rateList)
    #barGh.render_to_file("Dice6.svg")
    #
    # ---------------------------------------
    # barGh.add("D6", rateList)其中的D6所代表的
    # 視為圖例，也就是指用rateList中所有1-6點數各
    # 所出現次數的長條視圖
    #
# ----------------------------------------------
#
rollDice()
```

算字數畫圖

```python
with open(file="英文新聞.txt",mode="rt") as inpf :
    news = inpf.read()
    news = news.replace("\n","") #自己令一個
print(news)
```

```python
a=news.split()
d=dict()
for w in a:
    if d.get(w) == None :
        d[w] = 1
    else :
        d[w] = d.get(w) + 1
print(sorted(d.items(),key=lambda x:x[1], reverse=True)[1:20])  
sorted_count=sorted(d.items(),key=lambda x:x[1], reverse=True)[1:20]
print(sorted_count[6][0])
```

```python
plt.title("Word Count",color="r")
plt.xlabel("Word",color="r")
plt.ylabel("Count",color="r") 

x=[sorted_count[i][0] for i in range(19)]
y=[sorted_count[i][1] for i in range(19)]

plt.bar(x, y,color = 'r')
plt.xticks(rotation=90)
plt.show()
```

```python
def findSubStr(sentence,sub) :
    count=0
    n=0
    while sentence.find(sub,n) != -1 :
        if sentence.find(sub,n) != -1 :
            count += 1 
            n = sentence.find(sub,n) + 1
    return count
```

```python
with open(file="news20190405.txt",mode="rt") as inpf :
    news = inpf.read()
    news = news.replace("\n","") #
print(news)
```

```python
keywords = ['英特爾','高通','蘋果','台積電','三星']
d = dict()
for w in keywords :
    if d.get(w) == None :
        d[w] = findSubStr(news,w)
print(sorted(d.items(), key=lambda x:x[1], reverse=True))
```

```python
sorted_count=sorted(d.items(), key=lambda x:x[1], reverse=True)

plt.title("Word Count",color="r")
plt.xlabel("Word",color="r")
plt.ylabel("Count",color="r") 

x=[sorted_count[i][0] for i in range(4)]
y=[sorted_count[i][1] for i in range(4)]
plt.xticks(fontproperties=fontset, rotation=45)
plt.bar(x, y,color = 'r')
plt.xticks(rotation=90)
plt.show()
```
